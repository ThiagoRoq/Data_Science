```python
import pandas as pd
```

```python
dados = pd.read_csv("dataset/aluguel.csv",sep=';')
```

```python
dados
```

```python
dados["Tipo"].value_counts()
```

```python
bairros = dados["Tipo"].unique()
bairros
```

```python
df=[]
contagem=[]
```

```python
for bairro in bairros:
    count=0
    for contabilizar_bairros in dados["Tipo"]:
        if bairro==contabilizar_bairros:
            count=count+1
    contagem.append(count)
```

```python
df = pd.DataFrame(df)
df["Tipo"]=bairros
df["Quantidade"]=contagem
```
